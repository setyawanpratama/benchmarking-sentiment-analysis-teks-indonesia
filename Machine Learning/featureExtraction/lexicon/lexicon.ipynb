```python
import numpy as np
import pandas as pd

LEXICON_LIST = ["ID-OpinionWords", "InSet", "vania"]

pos = pd.read_csv("Lexicon/vania/positive.txt", header=None, names=['pos'])
list_pos = pos['pos'].tolist()
neg = pd.read_csv("Lexicon/vania/negative.txt", header=None, names=['neg'])
list_neg = neg['neg'].tolist()

df = pd.read_csv("C:/Users/Samuel Samosir/Documents/Python/Kuliah/TA/benchmarking-sentiment-analysis-teks-indonesia/Dataset/clean/cleaned_posneg_Dataset_A_HT_200.csv", delimiter=";", low_memory=False, header=0)
df
# new_header = df.iloc[0] #grab the first row for the header
# df = df[1:] #take the data less the header row
# df.columns = new_header #set the header row as the df header
```

```python
emosi = ["positif", "negatif"]
fitur_sentimen_all = []

for tweet in df["teks"].tolist():
    # inisiasi value
    value = [0, 0]
    emosi_value = {}
    for i in range(len(emosi)):
        emosi_value[emosi[i]] = value[i]

    list_kata = eval(tweet)
    for k in list_kata:
        if k in list_pos:
            emosi_value["positif"] += 1
        if k in list_neg:
            emosi_value["negatif"] += 1

    fitur_sentimen_perkalimat = list(emosi_value.values())
    fitur_sentimen_all.append(fitur_sentimen_perkalimat)

fitur_sentimen_all, emosi
```

```python
df_hasil = pd.DataFrame(fitur_sentimen_all, columns=emosi)
df_hasil
```
